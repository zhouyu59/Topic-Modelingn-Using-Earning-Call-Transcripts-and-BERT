# **1. Calculate relative frequency of topics**

Relative frequency (in %) is calculated by $ \frac {Number\;of \;Sentences\;Contain\;a\; Topic}{Total\;Number\;of\;Sentences}*100$



```python
topics = pd.read_csv("Downloads/BERT_topic_count_all_topics_five_years.csv")
```

```python
topics
```

```python
for column in topics:

  topics[column]=topics[column]/(np.sum(topics[column]))*100
```

```python
topics
```

Order the columns from 2018 Q1 to 2022 Q4

```python
change_column = ['2018_Q1',"2018_Q2","2018_Q3","2018_Q4","2019_Q1","2019_Q2","2019_Q3","2019_Q4","2020_Q1","2020_Q2","2020_Q3","2020_Q4","2021_Q1","2021_Q2","2021_Q3","2021_Q4","2022_Q1","2022_Q2","2022_Q3","2022_Q4"]
topics = topics.reindex(columns=change_column)
```
