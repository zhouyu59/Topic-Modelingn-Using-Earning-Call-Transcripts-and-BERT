# **2. BERTopic**

The following code will use BERTopic to find topics in all earning call transcripts and then count the number of earning call transcripts that contain the topics in each quarter.

```python
from bertopic import BERTopic
from cuml.cluster import HDBSCAN
from cuml.manifold import UMAP

hdbscan_model = HDBSCAN(min_samples=10, gen_min_span_tree=True, prediction_data=True) # import HDBSCAN clustering method that uses GPU to speed up processing
umap_model = UMAP(n_components=5, n_neighbors=15, min_dist=0.0) # import UMAP dimensionality reduction method that uses GPU to speed up processing
```

```python

topic_model = BERTopic(nr_topics="auto",umap_model=umap_model,hdbscan_model=hdbscan_model,verbose=True) # run BERTopic method using all earning call transcripts
topics, probs = topic_model.fit_transform(all_transcripts)
topic_model.get_topic_info()
```

use get_topics() method to ouput the topics that are identified by the model

The output will be a **dictionary**

The **keys** order from the most frequent topic to the least frequent topic
The **value** of the key is a list of words that constitute the topic

```python
topic_model.get_topics() # use get_topics method to know the topics that are identified
```
